# 男女身高体重鞋码数据可视化

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'png'
```

gender.xlsx 属性:

    1.height in cm
    2.weight in kg
    3.size of feet in 中国鞋码

```python
colume_names = ['index','gender','height','weight','size']
df= pd.read_excel('data/gender.xlsx',index_col=0,names=colume_names)
df.head(5)
```

```python
# 性别数据转换
df.replace('男','male',inplace=True)
df.replace('女','female',inplace=True)
df.head(5)
```

```python
# 男女比例
ax = df['gender'].value_counts().plot(kind = 'bar',figsize = (12,8),fontsize=15,rot = 0)
# 设置x,y轴标签
ax.set_ylabel('Counts',fontsize = 15)
ax.set_xlabel('gender',fontsize = 15)

plt.show()
```

```python
male_df = df.loc[df['gender']=='male']
female_df = df.loc[df['gender']=='female']
female_df.head(5)
```

```python
# 性别和身高、体重、鞋码的相关性

ax = male_df['height'].plot(kind = 'density',figsize = (12,8),fontsize=15)
female_df['height'].plot(kind = 'density',figsize = (12,8),fontsize=15)

ax.legend(['male','female'],fontsize = 12)
ax.set_ylabel('Density',fontsize = 15)
ax.set_xlabel('height',fontsize = 15)

plt.show()
```

```python
ax = male_df['weight'].plot(kind = 'density',figsize = (12,8),fontsize=15)
female_df['weight'].plot(kind = 'density',figsize = (12,8),fontsize=15)

ax.legend(['male','female'],fontsize = 12)
ax.set_ylabel('Density',fontsize = 15)
ax.set_xlabel('weight',fontsize = 15)

plt.show()
```

```python
ax = male_df['size'].plot(kind = 'density',figsize = (12,8),fontsize=15)
female_df['size'].plot(kind = 'density',figsize = (12,8),fontsize=15)

ax.legend(['male','female'],fontsize = 12)
ax.set_ylabel('Density',fontsize = 15)
ax.set_xlabel('size',fontsize = 15)

plt.show()
```
